```python
import json
import pandas as pd
import datetime


import plotly.graph_objects as go
import plotly.express as px


players = []
with open('Players.json') as f:
    players = json.loads(f.read())

players = pd.DataFrame(players)

data = []
with open('2021REG.json') as f:
    data = json.loads(f.read())
    
df2021 = pd.DataFrame(data)

with open('2020REG.json') as f:
    data = json.loads(f.read())
df2021['Year'] = '2021'

df2020 = pd.DataFrame(data)
df2020['Year'] = '2020'
with open('2019REG.json') as f:
    data = json.loads(f.read())
    
df2019 = pd.DataFrame(data)
df2019['Year'] = '2019'

# df = pd.concat([df2019,df2020,df2021])
df = pd.concat([df2021, players])
df = df2021.merge(players, left_on='PlayerID', right_on='PlayerID')
# # df = df.reindex(df.PlayerID)
# # df = pd.concat([df2021, players], axis=1).reindex(df1.index)


# position = ["RB","WR","TE","QB","K","W","R","T","DEF",]
boolean_series = df.PositionCategory_x.isin(['OFF'])
# boolean_series = df.Position_x.isin(position)
df = df[boolean_series]
```

```python
pd.set_option('display.max_columns', None)
df['Height'] = (df['HeightFeet'] * 12) + df["HeightInches"]
# df['FantasyPointsYahoo_rank'] = df.groupby("Position_x")["FantasyPointsYahoo"].rank(na_option='bottom',pct=True)
df['FantasyPointsYahoo_rank'] = df.groupby("Position_x")["FantasyPointsYahoo"].rank(na_option='bottom',pct=False, ascending=False)
# df['FantasyPointsYahoo_rank'] = df['FantasyPointsYahoo'].rank(pct=True,na_option='bottom')
```
